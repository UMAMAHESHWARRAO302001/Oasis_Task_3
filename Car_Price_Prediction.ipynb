```python
from sklearn.preprocessing import LabelEncoder as LE 
```

```python
cols = ['fueltype','aspiration','doornumber', 'carbody','enginetype','drivewheel','cylindernumber','wheelbase','fuelsystem']
for i in cols:
    car[f"{i}_le"]=LE().fit_transform(car[i])
    car=car.drop(i,axis=1)
car.head()
```

```python
from sklearn.preprocessing import MinMaxScaler as mms
```

```python
scaler = mms(feature_range=(-1, 1))
```

```python
df1= pd.DataFrame(scaler.fit_transform(car), columns=car.columns)
df1
```

```python
X=df1.drop('price',axis=1)
Y=df1['price']
```

```python
X.head()
```

```python
Y
```

```python
import matplotlib.pyplot as plt
```

```python
for i in df1.columns:
    print(i)
    if i !='price':
        plt.scatter(df1[i],df1['price'])
        plt.xlabel(i)
        plt.ylabel('Price')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test=tts(X,Y,random_state=42,test_size=0.5)
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
model=LR()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
Y_pred = model.predict(X_test)
score=mse(Y_test,Y_pred)
```

```python
print(f"Mean squared error is: {score}")
score1=model.score(X_test,Y_test)
```

```python
print(f"Accuracy is {(score1*100).__round__(2)}%")
```
